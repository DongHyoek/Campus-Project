```python
X_new = pd.read_csv(os.path.abspath("../input")+'/hwang_selected_features_train.csv', encoding = 'utf-8')
X_te_new = pd.read_csv(os.path.abspath("../input")+'/hwang_selected_features_test.csv', encoding = 'utf-8')
target = pd.read_csv(os.path.abspath("../input")+'/y_train.csv', encoding = 'cp949').group
```

```python
lgbm_params= {'num_leaves': 35,
        'min_child_samples': 182,
        'min_child_weight': 26,
        'max_depth': 10,
        'learning_rate': 0.018668206381448577,
        'n_estimators': 361,
        'reg_lambda': 0.6690395135477213,
        'reg_alpha': 0.7723190057522233,
        'subsample': 0.8855706663296421}
```

```python
# 모델 설정
lgbm = LGBMClassifier(**lgbm_params,
                objective='multiclass',
                metric='multi_logloss',
                nthread=4,
                random_state = 0
                )

catboost = CatBoostClassifier(random_state=0,iterations=500,depth=4,bootstrap_type ='Bayesian')
models = [lgbm,catboost]
```

```python
# vecstack 사용법을 이해하면 이전 수업시간에 설명했던 그림이 이해가 된다.
from vecstack import stacking
from sklearn.metrics import log_loss

S_train, S_test = stacking(models,                     # list of models
                           X_new, target, X_te_new,   # data
                           regression=False,           # classification task (if you need 
                                                       #     regression - set to True)
                           needs_proba=True,          # predict class labels (if you need 
                                                       #     probabilities - set to True) 
                           metric=log_loss,      # metric: callable
                           n_folds=5,                  # number of folds
                           stratified=True,            # stratified split for folds
                           shuffle=True,               # shuffle the data => 먼저 셔플링 할 것인지
                           random_state=0,             # ensure reproducibility
                           verbose=2)                  # print all info => 하는 과정을 추적할 것인지

# # 결과값이 S_train, S_test로 2개가 나옴
# meta_model = GBM.fit(S_train, y_train)
# accuracy_score(y_test, meta_model.predict(S_test))
```

```python
meta_model = LGBMClassifier(random_state = 0,n_estimators = 200,n_jobs = -1).fit(S_train,target)
test_preds = meta_model.predict_proba(S_test)
```

```python
pred_final = pd.DataFrame(test_preds).rename(columns = {0:'F20',1:'F30',2:'F40',3:'F50',4:'M20',5:'M30',6:'M40',7:'M50'})
pred_final
```

```python
test = pd.read_csv(os.path.abspath("../input")+'/X_test.csv',encoding = 'cp949')
tst_id = test['custid'].unique()
```

```python
submission = pd.concat([pd.DataFrame({'ID':tst_id}),pred_final],axis = 1)
```

```python
submission.to_csv('hwang_stacking.csv',index = False,encoding = 'utf-8')
```
